```python
#Importamos LIbrerías

import numpy as np
import tensorflow as tf
```

```python
#Variables
celsius = np.array([-35,-10,0,8,15,22,40], dtype=float)
fahrenheit = np.array([-31,14,32,46.4,59,71.6,104], dtype=float)
```

```python
#capa = tf.keras.layers.Dense(units=1,input_shape=[1])
#modelo= tf.keras.Sequential([capa])

#Aumentando Capas
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)

modelo= tf.keras.Sequential([oculta1,oculta2,salida])
```

```python
#Compilador, evalue los errores
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
               )
```

```python
print("Entrenando...")

historial = modelo.fit(celsius, fahrenheit, epochs=200, verbose=False)
print("Modelo entrenado!")
```

```python
# Graficar la función de pérdida
import matplotlib.pyplot as plt

plt.xlabel("Epochs")
plt.ylabel("Magnitud de Pérdida")
plt.plot(historial.history["loss"])
plt.show()
```

```python
print("Predicción")
resultado = modelo.predict([30.0])
print("El resultado es "+str(resultado)+ " f")
```
